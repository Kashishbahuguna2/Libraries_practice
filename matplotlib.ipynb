```python

import matplotlib.pyplot as plt 

x =[1,2,3,4,5,6,7]
y = [1000 , 5000,3500,2000,8000,4000,6000]

plt.plot(x,y , color = "blue" , linestyle = "--" , linewidth = 2 , marker = "o", label = "sale data" ) 
plt.xlabel("month")
plt.ylabel("sales amount")
plt.legend() 
plt.grid( color = "grey" , linestyle = ":" )
plt.xlim(1,7)
plt.ylim(1000 , 8000)
plt.xticks([1,2,3,4,5,6,7] , ['m1', 'm2', 'm3', 'm4', 'm5', 'm6', 'm7', ])

plt.title("Montly sales data ")
plt.show()
```

```python
# bar graph (use for comparison  between two things )
import matplotlib.pyplot  as plt 

product = ['m1' ,'m2' ,'m3' ,'m4' ,'m5' ,'m6' , 'm7' ]
sales = [1000,4000,3000,7000,4500,2000,8000]

plt.bar(product , sales , color = "purple" , label = "sales 2025")
plt.xlabel("product")
plt.ylabel("sales")
plt.grid(color = "grey" , linestyle = ":")
plt.title("product- sales data " , color = "purple")
plt.legend()

plt.show()
```

```python
# pie chart (percentagewise data )

import matplotlib.pyplot as plt 

region = ["north " , "south" , "east" , "west"]
sales = [ 20000 , 40000, 67000 , 90000 ]

plt.pie(sales , labels=  region , autopct = "%1.1f%%" )
plt.title("sales contribution by region")
plt.legend()
plt.show()
```

```python
# histogram ( use for seeinfg frequency)
import matplotlib.pyplot as plt 

marks = [35 , 50 , 60 , 7 , 80 ,78, 65, 89, 40, 96, 99, 87, 90 , 76, 70, 79, 59, 80, 85, 88]
plt.hist(marks , bins = 6 , color = "purple" , edgecolor = 'black')
plt.xlabel("marks")
plt.ylabel("no of student")
plt.title("mark distribution of student")
plt.show()
```

```python
import matplotlib.pyplot as plt 

hour = [1,2,3,4,5,6,7]
score = [40 , 60 , 70 , 80 ,90 ,95 , 99]

plt.scatter(hour, score  , marker = "^"  , color = "red" , label = "student dats")
plt.xlabel("no of hours study")
plt.ylabel("score in exams")
plt.title("score in exams on basis of no of hour study")
plt.legend()
plt.grid(color = "brown" , linestyle = ":")
plt.show()
```

```python
import matplotlib.pyplot as plt 

fig , ax = plt.subplots(3,1 , figsize = (20,25))

x = ["jan" , "feb" , "march" , "april" , "may" ,"june" , "july", "august" , "september" , "october" , "november" , "december"]
y = [20000 , 18000 , 19000, 15000 , 17000, 19500, 10000, 15000, 25000, 16000, 30000 , 29000]

ax[0].plot(x,y , color = "purple" , label = "sales data")
ax[0].set_title("line plot" , color = "brown")
ax[0].grid(linestyle = ":")
ax[0].legend()
ax[0].set_xlabel("months")
ax[0].set_ylabel("sales")

ax[1].bar(x,y , color = "red" ,  label = "sales data")
ax[1].set_title("bar plot" , color = "brown")
ax[1].grid(linestyle = ":")
ax[1].legend()
ax[1].set_xlabel("months")
ax[1].set_ylabel("sales")

ax[2].scatter(x,y , color = "orange" , marker = "^" , label ="sales data")
ax[2].set_title("scatter plot" , color = "brown")
ax[2].grid(linestyle = ":")
ax[2].legend()
ax[2].set_xlabel("months")
ax[2].set_ylabel("sales")

fig.suptitle("Sales data comparison" , color = "black" , fontsize = 19)

plt.tight_layout()

plt.savefig("chart.png" , dpi = 300 , bbox_inches = "tight")

plt.show()
```
