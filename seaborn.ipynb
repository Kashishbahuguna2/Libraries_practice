```python
import seaborn as sns
df = sns.load_dataset("titanic")
print(df.head())
```

```python
sns.scatterplot(x = "age" , y = "fare" , data = df , hue = "class" , palette = ['blue' , 'red' ,'green'])
```

```python
sns.histplot(df['age'] , kde = True , bins = 20)
```

```python
sns.barplot(x = 'sex' , y = 'fare' , data = df , palette = ['red' , 'blue' , 'green'] , hue = 'class')
```

```python
sns.boxplot(x = df['pclass'] , y = df['fare'] , hue = df['sex'] , palette = ['red' , 'blue'])
```

```python
sns.pairplot(df[['pclass' , 'survived' , 'fare' , 'sibsp' ] ], hue = 'sibsp')
```
